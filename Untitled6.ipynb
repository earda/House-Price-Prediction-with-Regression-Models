```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import Lasso
from sklearn import metrics
%matplotlib inline
```

```python
dataset = pd.read_csv('proje1.csv',encoding='latin5',sep=";")

def intdegeremlak(x):
    if x == 'Daire':
        return 1
    else:
        return 0
def intdegeray(x):
    if x == 'Haziran':
        return 1
    elif x == 'Mayıs':
        return 2
    else :
        return 0


def mahalleint(x):
    if x == 'Esentepe':
        return 1
    elif x == 'Yakacık':
        return 3
    elif x == 'Soğanlık':
        return 2
    elif x == 'Rahmanlar':
        return 5
    elif x == 'Uğurmumcu':
        return 7
    elif x == 'Topselvi':
        return 8
    elif x == 'Dragos':
        return 4
    else :
        return 0
dataset['Emlak Tipi']=dataset['Emlak Tipi'].apply(intdegeremlak)

dataset['Ay']=dataset['Ay'].apply(intdegeray)
dataset['Mahalle']=dataset['Mahalle'].apply(mahalleint)

dataset[150:200]
 
```

```python
def scatter_plot(feature,target):
    plt.figure(figsize=(16,8))
    plt.scatter(dataset[feature],
               dataset[target],
               c='black')
    plt.show
```

```python
scatter_plot('MetreKare','Fiyat')
scatter_plot('Oda Sayısı','Fiyat')
scatter_plot('Gün','Fiyat')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

Xs =dataset.drop(['Fiyat'], axis=1)
y = dataset['Fiyat'].values.reshape(-1,1)
lin_reg = LinearRegression()
MSEs = cross_val_score(lin_reg ,Xs,y ,scoring='neg_mean_squared_error',cv=5)
mean_MSE = np.mean(MSEs)
print(mean_MSE)
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge = Ridge()
parametreler = {'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20]}
ridge_regressor = GridSearchCV(ridge,parametreler,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(Xs,y)
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

lasso = Lasso()
parametreler = {'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20]}
lasso_regressor = GridSearchCV(lasso,parametreler,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(Xs,y)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
X, y = dataset.iloc[:,:-1],dataset.iloc[:,-1]
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
xg_reg = xgb.XGBRegressor(colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(X_train,y_train)

preds = xg_reg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```
